```python
import PIL
import numpy as np
from matplotlib import pyplot as plt


def show_images(iml, flip=True, columns=4, figsize=(17, 8)):
    
    iml = list(iml)
    
    plt.figure(figsize=figsize)
    for i, image in enumerate(iml):
        
        if flip and isinstance(image, np.ndarray):
            image = image[::-1]
        else:
            image = image.transform(PIL.Image.Transpose.FLIP_TOP_BOTTOM)
        
        plt.subplot(len(iml) // columns + 1, columns, i + 1)
        plt.axis('off')
        plt.imshow(image)
        

def show_image(image, flip=True):
    
    if isinstance(image, np.ndarray):
        image = PIL.Image.fromarray(image)
        
    if flip:
        image = image.transpose(PIL.Image.Transpose.FLIP_TOP_BOTTOM)
        
    return image.convert('RGB')
```

```python
import jupylet.rl as rl
```

```python
esquivar = rl.GameProcess('esquivar_jupylet')
```

```python
esquivar.start()
```

```python
o = esquivar.observe()
show_image(o['screen0'])
```

```python
games = rl.Games(['esquivar_jupylet'] * 8)
```

```python
games.start()
```

```python
show_images(o['screen0'] for o in games.observe())
```

```python
player = np.random.randint(-1, 1, (len(games.games), 5))
```

```python
games.step(player)
```

```python
show_images(o['screen0'] for o in games.observe())
```

```python
%timeit -n200 games.step(player)
```
